```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'M'
    elif x in ['F', 'Femal', 'female']:
        return 'F'
    else:
        return 'NaN'
```

```python
df['gender'] = list(map(clean_gender, df['gender']))
df['gender'].value_counts()
```

```python
df['income'] = df['income'].replace(0, np.nan)
df
```

```python
def rename_region(x):
    if x in ['California']:
        return 'West Region'
    elif x in ['Oregon']:
        return 'North West'
    elif x in ['Washington']:
        return 'East'
    elif x in ['Arizona','Nevada']:
        return 'Central'
    else:
        return 'NaN'
```

```python
df['st'] = list(map(rename_region, df['st']))
df['st'].value_counts()
```
